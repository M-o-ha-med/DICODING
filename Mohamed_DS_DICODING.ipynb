### Assessing Data

```python
data_hour = pd.read_csv("/tmp/hour.csv")
```

```python
data_hour.describe()
```

```python
data_hour['hr'] = data_hour['hr'].replace(to_replace = [x for x in range(0,24)] , value = [y for y in range(1,25)])
data_hour.head(10)
```

```python
data_hour.isna().sum()
```

### Cleaning Data

```python
data_hour["dteday"] = pd.to_datetime(data_hour["dteday"])
try:
  data_hour = data_hour.drop("instant" , axis = 1)
except:

  data_hour.info()
```

## Exploratory Data Analysis (EDA)

### Explore ...

```python
#pertanyaan 1
data_hour.groupby(by = "weathersit").agg(
{
    "cnt":["min" , "max","mean"],
}
)
```

```python
#pertanyaan 2
data_hour.groupby(by = "weathersit").casual.sum()
```

```python
data_hour.groupby(by = "season").casual.sum()
```

```python
#pertanyaan 3
print("jumlah sewa sepeda pada hari biasa\n")
data_hour.groupby(by = "weekday").agg(
{
  "casual":["min" , "max" , "mean"],
  "registered":["min" , "max" , "mean"] ,
  "cnt":["min" , "max" , "mean"]

}

).sum()
```

```python
print("jumlah sewa sepeda pada hari kerja\n")
data_hour.groupby(by = "holiday").agg(
{
  "casual":["min" , "max" , "mean"],
  "registered":["min" , "max" , "mean"] ,
  "cnt":["min" , "max" , "mean"]

}
).sum()
```

```python
print("jumlah sewa sepeda pada hari libur\n")
data_hour.groupby(by = "workingday").agg(
{
  "casual":["min" , "max" , "mean"],
  "registered":["min" , "max" , "mean"] ,
  "cnt":["min" , "max" , "mean"]

}

).sum()
```

```python
#pertanyaan 4
data_1 = data_hour["casual"]
data_2 = data_hour["registered"]
print("-----Casual-----")
print("min_casual   : " , data_1.min() )
print("max_casual   : " , data_1.max() )
print("mean_casual  : " , data_1.mean())
print("Sum_casual   : " , data_1.sum() ,"\n")
print("-----Registered-----")
print("min_casual   : " , data_2.min() )
print("max_casual   : " , data_2.max() )
print("mean_casual  : " , data_2.mean() )
print("Sum_casual   : " , data_2.sum() )

```

```python
#pertanyaan 5
data_hr_max = data_hour[data_hour.hr == data_hour.hr.max()]
data_hr_min = data_hour[data_hour.hr == data_hour.hr.min()]
data_hr_max.groupby("hr").agg({"cnt":["min" , "max" , "mean"] , "casual":["min" , "max" , "mean"] , "registered":["min" , "max" , "mean"] })
```

```python
data_hr_min.groupby("hr").agg({"cnt":["min" , "max" , "mean"] , "casual":["min" , "max" , "mean"] , "registered":["min" , "max" , "mean"] })
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
#jumlah sewa secara keseluruhan
fig , axis  = plt.subplots(nrows = 1 , ncols  = 2 , figsize = (20 , 8))
sns.barplot(x = data_hour.season , y = data_hour.cnt , ax = axis[0] )
sns.barplot(x = data_hour.weathersit, y = data_hour.cnt , ax = axis[1] )
plt.ylim(0, 250)
plt.suptitle("Cuaca dan keadaan cuaca terhadap jumlah sewa sepeda")
plt.show()
```

### Pertanyaan 2:

```python
plt.bar(x = data_hour.hr ,height = data_hour.cnt)
plt.xlabel("waktu (jam)")
plt.ylabel("Jumlah sewa")
plt.show()
```

###Pertanyaan 3:

```python
figure,axis  = plt.subplots(nrows = 1 , ncols = 3 ,  figsize=(24, 8))
sns.lineplot(x  = data_hour.temp , y = data_hour.cnt ,ax= axis[0])
sns.lineplot(x  = data_hour.hum , y = data_hour.cnt ,ax= axis[1])
sns.lineplot(x  = data_hour.windspeed , y = data_hour.cnt ,ax= axis[2])
plt.ylim(0,500)
plt.suptitle("perbandingan antara temperatur , kelembapan dan kecepatan angin  terhadap jumlah sewa sepeda",fontsize = 10)
plt.show()
```

###Pertanyaan 4:

```python
#axist = [data_hour["weekday"].mean() , data_hour["holiday" ].mean(), data_hour["workingday" ].mean()]
fig  ,axis = plt.subplots(nrows = 1 , ncols = 3 , figsize = (15 , 8))
sns.barplot(x = data_hour["weekday"] , y = data_hour.cnt , ax = axis[0])
sns.barplot(x = data_hour["holiday"] , y = data_hour.cnt , ax = axis[1])
sns.barplot(x = data_hour["workingday"] , y = data_hour.cnt , ax = axis[2])
plt.suptitle("perbandingan antara hari biasa , hari libur dan hari kerja  terhadap jumlah sewa sepeda",fontsize = 10)
plt.show()
```
